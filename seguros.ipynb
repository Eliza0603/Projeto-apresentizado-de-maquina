```python
import pandas as pd

df = pd.read_csv("segurosdatabase.csv")


print("--- Estrutura da Base ---")
display(df.head())


print("\n--- Verificação de Valores Nulos ---")
print(df.isnull().sum())

print("\n--- Verificação de Duplicatas ---")
print(f"Total de linhas duplicadas: {df.duplicated().sum()}")


print("\n--- Resumo dos Dados ---")
display(df.describe())
```

 Graficos:


1- "Fumantes têm custos médicos mais altos?"

**Resposta:**

Para analisar essa questão, foi feita uma comparação entre os custos médicos (charges) de fumantes e não fumantes.
Observou-se que a média de custos dos fumantes é significativamente maior do que a dos não fumantes.
Além disso, os fumantes apresentam maior variação nos custos, indicando maior risco financeiro.
Isso mostra que o tabagismo é um dos principais fatores que influenciam o aumento dos gastos com saúde.

```python
import seaborn as sns
import matplotlib.pyplot as plt


sns.set_theme(style="whitegrid")
plt.figure(figsize=(9, 6))


ax = sns.boxplot(
    x='smoker',
    y='charges',
    data=df,
    hue='smoker',
    palette={'yes': '#ff4d4d', 'no': '#2ecc71'}, # Vermelho e Verde
    width=0.6,
    legend=False
)


plt.title('Distribuição de Custos Médicos: Fumantes vs. Não Fumantes', fontsize=15, fontweight='bold', pad=20)
plt.ylabel('Custos Médicos ($)', fontsize=12)
plt.xlabel('Status de Tabagismo', fontsize=12)


plt.xticks([0, 1], ['Fumante', 'Não Fumante'])


sns.despine(left=True)

plt.show()
```

 **Justificativa:**


 O boxplot foi utilizado pois permite comparar a distribuição dos custos entre dois grupos, evidenciando diferenças de mediana, dispersão e presença de outliers.

4 - "Ter filhos aumenta os custos médicos?"


**Resposta:**

 Para analisar essa questão, foi feita uma comparação entre o número de filhos (children) e os custos médicos (charges).
Observa-se que não há um aumento significativo nos custos conforme o número de filhos cresce.
As médias de custos permanecem relativamente semelhantes entre os grupos com diferentes quantidades de filhos.
Isso indica que o número de dependentes não é um fator determinante para o aumento dos gastos médicos individuais.

```python
import seaborn as sns
import matplotlib.pyplot as plt


df_pt = df.copy()
df_pt['children_pt'] = df_pt['children'].map({
    0: '0 Filhos',
    1: '1 Filho',
    2: '2 Filhos',
    3: '3 Filhos',
    4: '4 Filhos',
    5: '5 Filhos'
})


sns.set_theme(style="whitegrid")
plt.figure(figsize=(10, 6))


ax = sns.barplot(
    x='children_pt',
    y='charges',
    data=df_pt,
    hue='children_pt',
    palette='mako',
    legend=False,
    order=['0 Filhos', '1 Filho', '2 Filhos', '3 Filhos', '4 Filhos', '5 Filhos']
)


plt.title('Custos Médicos por Quantidade de Filhos', fontsize=16, fontweight='bold', pad=20)
plt.xlabel('Número de Dependentes (Filhos)', fontsize=12)
plt.ylabel('Custos Médicos Médios ($)', fontsize=12)

sns.despine(left=True, bottom=True)

plt.show()
```

**Justificativa:**

O gráfico de barras foi utilizado pois permite comparar a média dos custos médicos entre diferentes categorias (quantidade de filhos), facilitando a identificação de possíveis variações entre os grupos.

5- "A região geográfica influencia os custos do seguro de saúde?"


```python


import pandas as pd
df = pd.read_csv("segurosdatabase.csv")
print(df['region'].unique())

import seaborn as sns
import matplotlib.pyplot as plt


traducao = {
    'southwest': 'Sudoeste',
    'southeast': 'Sudeste',
    'northwest': 'Noroeste',
    'northeast': 'Nordeste'
}
df['Regiao_Exibicao'] = df['region'].map(traducao)

sns.set_theme(style="whitegrid")
plt.figure(figsize=(10, 6))


ax = sns.barplot(
    x='Regiao_Exibicao',
    y='charges',
    data=df,
    hue='Regiao_Exibicao',
    palette='viridis',
    legend=False
)


plt.title('Média de Custos Médicos por Região', fontsize=15, fontweight='bold', pad=20)
plt.xlabel('Região Geográfica', fontsize=12)
plt.ylabel('Custos Médicos Médios ($)', fontsize=12)



sns.despine(left=True, bottom=True)
plt.show()
```

**Justificativa:**

 O gráfico de barras é adequado para comparar médias entre categorias, facilitando a visualização de diferenças entre regiões.

6 - "Qual a distribuição de idade dos segurados?"

**Resposta:**


Resposta:
A análise da distribuição etária revela que a base de clientes é composta por uma vasta gama de idades, com uma concentração notável de jovens adultos (em torno dos 18-20 anos). A presença equilibrada de diversas faixas etárias é fundamental para a seguradora, pois permite diluir o risco atuarial. No entanto, o volume de jovens pode indicar um grupo com menor histórico de saúde, enquanto a distribuição contínua até os 64 anos sugere uma base madura que tende a ter custos mais previsíveis, porém potencialmente elevados.

```python
import seaborn as sns
import matplotlib.pyplot as plt


sns.set_theme(style="white")
plt.figure(figsize=(10, 6))


ax = sns.histplot(
    df['age'],
    bins=20,
    kde=True,
    color='#440154',
    line_kws={'linewidth': 3}
)


plt.title('Distribuição de Idade dos Segurados', fontsize=16, fontweight='bold', pad=20)
plt.xlabel('Idade (Anos)', fontsize=12)
plt.ylabel('Frequência (Quantidade de Pessoas)', fontsize=12)


sns.despine()

plt.show()
```

**Justificativa:**

O Histograma foi selecionado por ser a ferramenta estatística mais eficiente para representar a distribuição de uma variável numérica contínua (idade). Diferente de um gráfico de barras comum, ele agrupa os dados em intervalos (bins), permitindo identificar visualmente a densidade da população, a presença de modas (picos de frequência) e a simetria dos dados. A inclusão da linha de estimativa de densidade de kernel (KDE) auxilia na percepção da forma da distribuição, facilitando a identificação de tendências que orientam o cálculo de risco do seguro.

7- "Existe interação entre fumar + IMC nos custos?"

 **Resposta:**

 Para analisar essa interação, foi observada a relação entre IMC (bmi) e custos médicos (charges), diferenciando fumantes e não fumantes.
Observa-se que indivíduos fumantes com IMC elevado apresentam os maiores custos médicos.
Isso indica que a combinação entre tabagismo e alto IMC potencializa significativamente os gastos com saúde, evidenciando uma interação entre essas variáveis.

```python
import seaborn as sns
import matplotlib.pyplot as plt


df_pt = df.copy()
df_pt['smoker'] = df_pt['smoker'].map({'yes': 'Fumante', 'no': 'Não Fumante'})


sns.set_theme(style="ticks")
plt.figure(figsize=(10, 6))


sns.scatterplot(
    x='bmi',
    y='charges',
    hue='smoker',
    data=df_pt,
    palette={'Fumante': '#E74C3C', 'Não Fumante': '#2ECC71'},
    alpha=0.7,
    s=60,
    edgecolor='w'
)


plt.title('Relação entre IMC e Custos Médicos', fontsize=16, fontweight='bold', pad=15)
plt.xlabel('IMC (Índice de Massa Corporal)', fontsize=12)
plt.ylabel('Custos Médicos ($)', fontsize=12)


plt.legend(title='Status de Tabagismo', title_fontsize='11', loc='upper left')


sns.despine()

plt.grid(True, linestyle='--', alpha=0.5)
```

**Justificativa:**  
O gráfico de dispersão com diferenciação por cor permite visualizar simultaneamente três variáveis, sendo ideal para identificar interações entre elas.

8-" Quais variáveis têm maior influência sobre os custos médicos?"




**Resposta:**

A análise final revela que o custo médico não é definido por uma única variável, mas pela combinação estratégica de fatores. O gráfico de dispersão abaixo demonstra que:

Não Fumantes (Verde): Mantêm custos baixos (abaixo de $15.000) independentemente do IMC, com um crescimento muito sutil conforme a idade ou peso aumentam.

Fumantes (Vermelho): São divididos em dois grupos críticos. Aqueles com IMC abaixo de 30 já gastam o dobro dos não fumantes. No entanto, ao ultrapassar a marca de 30 de IMC (obesidade), os custos saltam para a faixa de $30.000 a $60.000.
Portanto, a interação entre Tabagismo e IMC elevado é o principal preditor de alto risco financeiro para a seguradora.

```python

import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(style="whitegrid")
plt.figure(figsize=(11, 7))


ax = sns.scatterplot(
    x='bmi',
    y='charges',
    hue='smoker',
    data=df,
    palette={'yes': '#ff4d4d', 'no': '#2ecc71'},
    s=80,
    alpha=0.7,
    edgecolor='w'
)


plt.legend(
    title='Fumante?',
    labels=['Sim', 'Não'],
    loc='upper left',
    title_fontsize='12',
    fontsize='10'
)


plt.title('Influência Combinada: IMC e Tabagismo nos Custos', fontsize=15, fontweight='bold', pad=20)
plt.xlabel('Índice de Massa Corporal (IMC)', fontsize=12)
plt.ylabel('Custos Médicos ($)', fontsize=12)

sns.despine()
plt.show()
```

**Justificativa:**  

Esta visualização utiliza um Gráfico de Dispersão Multivariado, que é a ferramenta mais sofisticada da Análise Exploratória para identificar clusters (agrupamentos) e interações não lineares. Ao cruzar os dados de IMC (numérico) com o status de fumante (categórico), conseguimos provar que o impacto do IMC não é constante, mas sim condicionado ao comportamento do segurado. A inclusão da linha vertical no IMC 30 serve como um suporte visual para a tomada de decisão de negócio, segmentando os clientes em faixas de risco claras.
